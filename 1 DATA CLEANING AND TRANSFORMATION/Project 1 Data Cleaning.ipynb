## b) Now it's time to check the type of each column to see if there are inconsistencies eg. year saved as float

With the dtypes attribute we can see there are some issues that need to be adressed.
*   contract_valid_until should be str not Float, the same for league_rank.
*    release_clause_eur, team_jersey_num, nation_jersey_num,league_rank they should be Int64


```python
pd.options.display.max_columns = None
pd.set_option('display.max_rows', 200)
df.dtypes
```

## c) Now we are going to fix this types issues



```python
df['contract_valid_until'] = df['contract_valid_until'].astype('Int64').astype('str')
#first we convert to integer then to str to avoid decimals
df.astype
```

```python
df.head(3)
```

```python
df['release_clause_eur'] = df['release_clause_eur'].astype('Int64')
```

```python
df['team_jersey_number'] = df['team_jersey_number'].astype('Int64')
```

```python
df['nation_jersey_number'] = df['nation_jersey_number'].astype('Int64')
```

```python
df.head(10)
```

```python
df['league_rank'] = df['league_rank'].astype('Int64').astype('str')
```

```python
df.dtypes
```

## d) We can work some magic on the joined column to have day, month and year in their own columns


```python
df['joined_converted'] = pd.to_datetime(df['joined'], infer_datetime_format=True)
```

```python
df.head(1)
df.asty
```

```python
df['joined_year']= df['joined_converted'].dt.year
df['joined_year']=df['joined_year'].astype ('Int64')
```

```python
df.head(1)
```

```python
df['joined_month']= df['joined_converted'].dt.month
df['joined_month']=df['joined_month'].astype ('Int64')
```

```python
df['joined_day']= df['joined_converted'].dt.day
df['joined_day']=df['joined_day'].astype ('Int64')
```
