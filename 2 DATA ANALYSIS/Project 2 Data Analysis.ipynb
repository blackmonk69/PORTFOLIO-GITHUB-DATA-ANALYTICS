## b) Supose we need to find out more about years coding replies, say which period of time has more ocurrence in the survey

We do an aggregate function grouping YearsCode and convert de resulting groupby object into a dataframe

```python
df3 = df.groupby(['YearsCode'])['YearsCode'].count()
df4= pd.DataFrame(df3)
df4.head()
```

```python
df4.columns
```

## c) Let's add a percentage column which will represent the impact of the count column on the total


```python
df4['percentage']=df4['YearsCode']/df4['YearsCode'].sum() *100
#removing the index heading YearsCode
df4.index.name = None
df4.reset_index(inplace=True)
df4.head()
```

## d) Some column Renaming...

```python
df4.rename(columns = {'index':'QTY_Years_coding'},inplace=True)
df4.rename(columns = {'YearsCode':'QTY_Ocurrencies'},inplace=True)
```

## e) Some Column sorting...


```python
df4=df4.sort_values('percentage',ascending=False)
df4.reset_index(inplace=True)
```

```python
df4.head()
```

## g) Is there a relation between coding experience and the level of pay?

###  The last column shows the annual compensation converted to USD, so we can see the variation of it grouping by Years Coding

```python
df_compensation_vs_yearsCoding = df.groupby(['YearsCodePro'])['ConvertedCompYearly'].mean()
df_compensation_vs_yearsCoding=pd.DataFrame(df_compensation_vs_yearsCoding)
df_compensation_vs_yearsCoding.rename(columns = {'ConvertedCompYearly':'Mean_Yearly_Salary'},inplace=True)
df_compensation_vs_yearsCoding=df_compensation_vs_yearsCoding.sort_values('Mean_Yearly_Salary',ascending=False)

df_compensation_vs_yearsCoding["Mean_Yearly_Salary"]=df_compensation_vs_yearsCoding["Mean_Yearly_Salary"].round(2)
df_compensation_vs_yearsCoding
```

## I) What if we would like to know if we are more likely to get a job as a developer if we have a Master's Degree

```python
#first we group by EdLevel
df_group = df.groupby("EdLevel")
#once we have this grouping we pick the group with Master's degree
df_group=df_group.get_group('Master’s degree (M.A., M.S., M.Eng., MBA, etc.)')
df_group=pd.DataFrame(df_group)
# We again apply another grouping operation this time considering MainBranch to know the figures for every line
# and we count the ocurrencies of every Main Branch 
df_group.groupby("MainBranch")['MainBranch'].count()
#Conclusion: within the group of people with a Master's degree the proportion of Developers outnumbers the other items like
# `learning to code` or 'coder for hobby'
```
